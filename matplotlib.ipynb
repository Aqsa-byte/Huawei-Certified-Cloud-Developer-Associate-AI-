```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/Titanicdataset.csv')
df.head()
```

```python
df.tail()
```

```python
sex_counts = df['Sex'].value_counts()

plt.figure(figsize=(6,4))
plt.bar(sex_counts.index, sex_counts.values, color=['skyblue', 'lightcoral'])
plt.title('Passenger Count by Gender')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.scatter(df['Age'], df['Fare'], alpha=0.6, c='purple')
plt.title('Scatter Plot: Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.hist(df['Age'].dropna(), bins=20, color='green', edgecolor='black')
plt.title('Histogram of Passenger Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df['Fare'], color='orange')
plt.title('Line Plot: Fare across Passengers')
plt.xlabel('Passenger Index')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

```python
pclass_counts = df['Pclass'].value_counts()
labels = ['Class 1', 'Class 2', 'Class 3']

plt.figure(figsize=(6,6))
plt.pie(pclass_counts, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Passenger Class Distribution')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(7,5))
df.boxplot(column='Age', by='Sex')
plt.title('Box Plot of Age by Gender')
plt.suptitle('')  # Remove default title
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()
```
